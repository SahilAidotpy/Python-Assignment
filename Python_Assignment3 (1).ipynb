```python
import math

# ---------------- Q1 ----------------
# Generate primes less than 100 and arrange in spiral format

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

# Generate primes less than 100
primes = [i for i in range(2, 100) if is_prime(i)]

# Spiral matrix size (smallest square that fits all primes)
n = math.ceil(math.sqrt(len(primes)))
spiral = [[0]*n for _ in range(n)]

# Directions: right, down, left, up
dirs = [(0,1), (1,0), (0,-1), (-1,0)]
x, y, d = 0, 0, 0

for prime in primes:
    spiral[x][y] = prime
    nx, ny = x + dirs[d][0], y + dirs[d][1]
    if not (0 <= nx < n and 0 <= ny < n and spiral[nx][ny] == 0):
        d = (d + 1) % 4
        nx, ny = x + dirs[d][0], y + dirs[d][1]
    x, y = nx, ny

print("Prime Spiral (<100):")
for row in spiral:
    print(" ".join(f"{val:2d}" if val != 0 else "  " for val in row))
```

```python
# ---------------- Q2 ----------------
# Text-based graph of hourly views

def text_graph(views):
    print("\nHourly Views Graph:")
    for hour, v in enumerate(views):
        stars = "*" * (v // 5)
        print(f"Hour {hour:2d}: {stars}")

# Example hourly views (replace with actual input)
hourly_views = [12, 25, 7, 0, 40, 15, 55, 32, 18, 5, 60, 20, 0, 35, 45, 10, 8, 28, 33, 14, 12, 25, 7, 0]
text_graph(hourly_views)
```

```python
# ---------------- Q3 ----------------
# Revenue calculation based on total views

def calculate_revenue(views):
    total_views = sum(views)
    revenue = 0
    if total_views <= 10:
        revenue = total_views * 0.50
    elif total_views <= 30:
        revenue = 10 * 0.50 + (total_views - 10) * 0.30
    else:
        revenue = 10 * 0.50 + 20 * 0.30 + (total_views - 30) * 0.10
    return revenue

daily_revenue = calculate_revenue(hourly_views)
print(f"\nTotal Daily Revenue: ${daily_revenue:.2f}")
```
